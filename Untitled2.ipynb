```python
import tensorflow as tf
from tensorflow.keras import datasets

# This one line replaces manual uploading and unzipping
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

print(f"Data Loaded: {train_images.shape[0]} training images, {test_images.shape[0]} test images")
```

```python
from tensorflow.keras import layers, models

# ENCODER: Compresses 32x32x3 -> 8x8x8 (512 features)
input_img = layers.Input(shape=(32, 32, 3))
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x)
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
encoded = layers.MaxPooling2D((2, 2), padding='same')(x)

# DECODER: Reconstructs 8x8x8 -> 32x32x3
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(encoded)
x = layers.UpSampling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)
decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = models.Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train it to reconstruct itself
autoencoder.fit(train_images, train_images, epochs=10, batch_size=128, validation_data=(test_images, test_images))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import time

# --- 1. Load and Preprocess Data ---
print("Step 1: Loading CIFAR-10...")
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
x_train, x_test = x_train.astype('float32') / 255.0, x_test.astype('float32') / 255.0

# Using a subset for SVM to ensure it runs quickly in Colab
svm_train_size = 5000
svm_test_size = 1000
y_train_sub, y_test_sub = y_train[:svm_train_size].ravel(), y_test[:svm_test_size].ravel()

# --- 2. Build and Train Autoencoder ---
print("Step 2: Training Autoencoder (Unsupervised Compression)...")
input_img = layers.Input(shape=(32, 32, 3))
# Encoder
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x)
encoded = layers.Conv2D(8, (3, 3), activation='relu', padding='same')(x) # Bottleneck
# Decoder
x = layers.Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
x = layers.UpSampling2D((2, 2))(x)
decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = models.Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(x_train, x_train, epochs=5, batch_size=128, verbose=0) # Epochs kept low for speed

# --- 3. Hybrid Feature Extraction ---
print("Step 3: Extracting Latent Features...")
encoder_model = models.Model(input_img, encoded)
train_features = encoder_model.predict(x_train[:svm_train_size]).reshape(svm_train_size, -1)
test_features = encoder_model.predict(x_test[:svm_test_size]).reshape(svm_test_size, -1)

# --- 4. Train Supervised Models ---
print("Step 4: Training SVM on Compressed Features...")
start = time.time()
svm_hybrid = SVC(kernel='linear')
svm_hybrid.fit(train_features, y_train_sub)
hybrid_acc = accuracy_score(y_test_sub, svm_hybrid.predict(test_features))
hybrid_time = time.time() - start

print("Step 5: Training Baseline SVM on Raw Pixels...")
x_train_raw = x_train[:svm_train_size].reshape(svm_train_size, -1)
x_test_raw = x_test[:svm_test_size].reshape(svm_test_size, -1)
start = time.time()
svm_raw = SVC(kernel='linear')
svm_raw.fit(x_train_raw, y_train_sub)
raw_acc = accuracy_score(y_test_sub, svm_raw.predict(x_test_raw))
raw_time = time.time() - start

# --- 5. Final Visualization & Comparison ---
print("\n" + "="*30)
print("FINAL ANALYSIS")
print("="*30)

# Feature Reduction Stats
orig_feats = 32*32*3
comp_feats = train_features.shape[1]
print(f"Raw Features: {orig_feats}")
print(f"Compressed Features: {comp_feats}")
print(f"Data Reduction: {((orig_feats - comp_feats)/orig_feats)*100:.2f}%")
print(f"Hybrid SVM Accuracy: {hybrid_acc*100:.2f}% (Time: {hybrid_time:.2f}s)")
print(f"Raw SVM Accuracy: {raw_acc*100:.2f}% (Time: {raw_time:.2f}s)")

# Visual Comparison of Images
reconstructed = autoencoder.predict(x_test[:10])
plt.figure(figsize=(15, 5))
for i in range(10):
    # Original
    plt.subplot(2, 10, i + 1)
    plt.imshow(x_test[i])
    plt.axis('off')
    if i == 0: plt.title("Originals")
    # Reconstructed
    plt.subplot(2, 10, i + 11)
    plt.imshow(reconstructed[i])
    plt.axis('off')
    if i == 0: plt.title("Reconstructed")
plt.show()
```
