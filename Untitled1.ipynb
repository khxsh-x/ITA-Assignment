```python
# ============================================================
# HYBRID MACHINE LEARNING FOR EARLY DISEASE DIAGNOSIS
# Breast Cancer Dataset - Google Colab Ready
# Algorithms: K-Means, Hierarchical Clustering, PCA,
#             SVM, Random Forest, Neural Network
# ============================================================

# ── STEP 1: Install & Import ─────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (classification_report, confusion_matrix,
                             accuracy_score, roc_auc_score, roc_curve)

# Unsupervised
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA

# Supervised
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neural_network import MLPClassifier

from scipy.cluster.hierarchy import dendrogram, linkage

print("✅ All libraries imported successfully!")

# ── STEP 2: Load & Explore Dataset ───────────────────────────
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='diagnosis')   # 0 = malignant, 1 = benign

print("\n📊 Dataset Overview")
print(f"   Samples  : {X.shape[0]}")
print(f"   Features : {X.shape[1]}")
print(f"   Malignant (0): {(y == 0).sum()}  |  Benign (1): {(y == 1).sum()}")
print("\nFirst 5 rows:")
display(X.head())


# ── STEP 3: Preprocessing ────────────────────────────────────
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
print(f"\n✅ Train size: {X_train.shape[0]}  |  Test size: {X_test.shape[0]}")

# ╔══════════════════════════════════════════════════════════╗
# ║        UNSUPERVISED LEARNING (Pattern Discovery)         ║
# ╚══════════════════════════════════════════════════════════╝

# ── STEP 4: PCA – Dimensionality Reduction ───────────────────
print("\n" + "="*55)
print("  UNSUPERVISED LEARNING")
print("="*55)

pca = PCA(n_components=10, random_state=42)
X_pca = pca.fit_transform(X_scaled)

explained_var = np.cumsum(pca.explained_variance_ratio_)
print(f"\n🔬 PCA: 2 components explain {explained_var[1]*100:.1f}% of variance")
print(f"       10 components explain {explained_var[9]*100:.1f}% of variance")

# Scree plot
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].bar(range(1, 11), pca.explained_variance_ratio_*100,
            color='steelblue', edgecolor='black')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance (%)')
axes[0].set_title('PCA – Scree Plot', fontweight='bold')

# 2D PCA scatter
X_pca2 = PCA(n_components=2, random_state=42).fit_transform(X_scaled)
colors = ['#E74C3C' if label == 0 else '#2ECC71' for label in y]
axes[1].scatter(X_pca2[:, 0], X_pca2[:, 1], c=colors, alpha=0.6, edgecolors='k', s=40)
axes[1].set_xlabel('PC 1')
axes[1].set_ylabel('PC 2')
axes[1].set_title('PCA – 2D Projection', fontweight='bold')
from matplotlib.patches import Patch
axes[1].legend(handles=[Patch(color='#E74C3C', label='Malignant'),
                         Patch(color='#2ECC71', label='Benign')])
plt.suptitle('Principal Component Analysis (PCA)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# ── PCA: Top 10 Feature Contributions ───────────────────────
pca_components = pd.DataFrame(
    pca.components_,
    columns=data.feature_names,
    index=[f'PC{i+1}' for i in range(10)]
)

# For each PC, find which original feature contributes most
print("\n🔍 Most influential original feature per Principal Component:")
for pc in pca_components.index:
    top_feature = pca_components.loc[pc].abs().idxmax()
    top_value   = pca_components.loc[pc].abs().max()
    print(f"   {pc}: {top_feature}  (loading = {top_value:.3f})")

# Heatmap of all feature contributions across all 10 PCs
fig, ax = plt.subplots(figsize=(18, 6))
sns.heatmap(
    pca_components.T,
    cmap='coolwarm',
    center=0,
    annot=False,
    linewidths=0.4,
    ax=ax
)
ax.set_title('PCA Component Loadings — Original Feature Contributions',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Principal Component')
ax.set_ylabel('Original Feature')
plt.tight_layout()
plt.show()

# Bar chart: which features contribute most to PC1 specifically
pc1_loadings = pca_components.loc['PC1'].abs().sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(10, 4))
pc1_loadings.plot(kind='bar', color='steelblue', edgecolor='black', ax=ax)
ax.set_title('Top 10 Original Features Contributing to PC1',
             fontweight='bold', fontsize=13)
ax.set_ylabel('Absolute Loading')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.tight_layout()
plt.show()


# ── STEP 5: K-Means Clustering ───────────────────────────────
print("\n--- K-Means Clustering ---")
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_pca2)

# Align cluster labels with actual classes
from scipy.stats import mode
def align_labels(pred, true):
    aligned = np.zeros_like(pred)
    for cluster in np.unique(pred):
        mask = pred == cluster
        aligned[mask] = mode(true[mask], keepdims=True)[0][0]
    return aligned

kmeans_aligned = align_labels(kmeans_labels, y.values)
kmeans_acc = accuracy_score(y, kmeans_aligned)
print(f"   K-Means Cluster Accuracy : {kmeans_acc*100:.1f}%")

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
scatter = axes[0].scatter(X_pca2[:, 0], X_pca2[:, 1],
                          c=kmeans_labels, cmap='coolwarm', alpha=0.6,
                          edgecolors='k', s=40)
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                marker='X', s=200, c='black', label='Centroids')
axes[0].set_title('K-Means Clustering', fontweight='bold')
axes[0].set_xlabel('PC 1'); axes[0].set_ylabel('PC 2')
axes[0].legend()

axes[1].scatter(X_pca2[:, 0], X_pca2[:, 1], c=colors, alpha=0.6,
                edgecolors='k', s=40)
axes[1].set_title('Actual Labels (Ground Truth)', fontweight='bold')
axes[1].set_xlabel('PC 1'); axes[1].set_ylabel('PC 2')
plt.suptitle('K-Means vs Actual Labels', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()



# ╔══════════════════════════════════════════════════════════╗
# ║         SUPERVISED LEARNING (Classification)            ║
# ╚══════════════════════════════════════════════════════════╝

print("\n" + "="*55)
print("  SUPERVISED LEARNING")
print("="*55)

# PCA-reduced features for supervised models (10 components)
X_train_pca = pca.transform(X_train)
X_test_pca  = pca.transform(X_test)

# ── STEP 7: Support Vector Machine ───────────────────────────
print("\n--- Support Vector Machine (SVM) ---")
svm = SVC(kernel='rbf', C=1.0, probability=True, random_state=42)
svm.fit(X_train_pca, y_train)
svm_pred = svm.predict(X_test_pca)
svm_acc  = accuracy_score(y_test, svm_pred)
svm_auc  = roc_auc_score(y_test, svm.predict_proba(X_test_pca)[:, 1])
print(f"   Accuracy : {svm_acc*100:.2f}%  |  ROC-AUC : {svm_auc:.4f}")
print(classification_report(y_test, svm_pred,
      target_names=['Malignant', 'Benign']))

# ── STEP 8: Random Forest ─────────────────────────────────────
print("\n--- Random Forest ---")
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_pca, y_train)
rf_pred = rf.predict(X_test_pca)
rf_acc  = accuracy_score(y_test, rf_pred)
rf_auc  = roc_auc_score(y_test, rf.predict_proba(X_test_pca)[:, 1])
print(f"   Accuracy : {rf_acc*100:.2f}%  |  ROC-AUC : {rf_auc:.4f}")
print(classification_report(y_test, rf_pred,
      target_names=['Malignant', 'Benign']))

# Feature importance
importances = rf.feature_importances_
pca_labels  = [f'PC{i+1}' for i in range(10)]
sorted_idx  = np.argsort(importances)[::-1]

fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(pca_labels, importances[sorted_idx], color='darkorange', edgecolor='black')
ax.set_title('Random Forest – Feature Importances (PCA Components)',
             fontweight='bold')
ax.set_ylabel('Importance')
plt.tight_layout()
plt.show()

# ── STEP 9: Neural Network (MLP) ─────────────────────────────
print("\n--- Neural Network (MLP / Deep Learning) ---")
nn = MLPClassifier(hidden_layer_sizes=(64, 32, 16),
                   activation='relu', max_iter=500,
                   random_state=42, early_stopping=True)
nn.fit(X_train_pca, y_train)
nn_pred = nn.predict(X_test_pca)
nn_acc  = accuracy_score(y_test, nn_pred)
nn_auc  = roc_auc_score(y_test, nn.predict_proba(X_test_pca)[:, 1])
print(f"   Accuracy : {nn_acc*100:.2f}%  |  ROC-AUC : {nn_auc:.4f}")
print(classification_report(y_test, nn_pred,
      target_names=['Malignant', 'Benign']))

# Loss curve
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(nn.loss_curve_, color='steelblue', linewidth=2)
ax.set_title('Neural Network – Training Loss Curve', fontweight='bold')
ax.set_xlabel('Iterations')
ax.set_ylabel('Loss')
plt.tight_layout()
plt.show()

# ── STEP 10: Hybrid Ensemble Model ───────────────────────────
print("\n" + "="*55)
print("  HYBRID MODEL (SVM + RF + NN Ensemble)")
print("="*55)

hybrid = VotingClassifier(
    estimators=[('svm', svm), ('rf', rf), ('nn', nn)],
    voting='soft'
)
hybrid.fit(X_train_pca, y_train)
hybrid_pred    = hybrid.predict(X_test_pca)
hybrid_acc     = accuracy_score(y_test, hybrid_pred)
hybrid_auc     = roc_auc_score(y_test, hybrid.predict_proba(X_test_pca)[:, 1])
hybrid_cv      = cross_val_score(hybrid, X_pca, y, cv=5, scoring='accuracy')

print(f"\n   Hybrid Accuracy   : {hybrid_acc*100:.2f}%")
print(f"   Hybrid ROC-AUC    : {hybrid_auc:.4f}")
print(f"   5-Fold CV Mean    : {hybrid_cv.mean()*100:.2f}% ± {hybrid_cv.std()*100:.2f}%")
print(classification_report(y_test, hybrid_pred,
      target_names=['Malignant', 'Benign']))

# ── STEP 11: Confusion Matrix – All Models ───────────────────
models      = [svm, rf, nn, hybrid]
model_names = ['SVM', 'Random Forest', 'Neural Network', 'Hybrid Ensemble']
preds       = [svm_pred, rf_pred, nn_pred, hybrid_pred]

fig, axes = plt.subplots(1, 4, figsize=(18, 4))
for ax, name, pred in zip(axes, model_names, preds):
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=['Malignant','Benign'],
                yticklabels=['Malignant','Benign'])
    ax.set_title(f'{name}\nAcc: {accuracy_score(y_test,pred)*100:.1f}%',
                 fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
plt.suptitle('Confusion Matrices – All Models', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# ── STEP 12: ROC Curves ──────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 6))
colors_roc = ['#3498DB','#E67E22','#9B59B6','#E74C3C']
for model, name, color in zip(models, model_names, colors_roc):
    proba = model.predict_proba(X_test_pca)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, proba)
    auc  = roc_auc_score(y_test, proba)
    ax.plot(fpr, tpr, color=color, linewidth=2,
            label=f'{name} (AUC = {auc:.3f})')
ax.plot([0,1],[0,1],'k--', linewidth=1)
ax.set_xlabel('False Positive Rate', fontsize=12)
ax.set_ylabel('True Positive Rate', fontsize=12)
ax.set_title('ROC Curves – All Models', fontsize=14, fontweight='bold')
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()

# ── STEP 13: Final Performance Summary ───────────────────────
print("\n" + "="*55)
print("  FINAL PERFORMANCE SUMMARY")
print("="*55)

summary = pd.DataFrame({
    'Model'    : model_names,
    'Type'     : ['Supervised','Supervised','Supervised','Hybrid'],
    'Accuracy' : [f"{accuracy_score(y_test,p)*100:.2f}%" for p in preds],
    'ROC-AUC'  : [f"{roc_auc_score(y_test, m.predict_proba(X_test_pca)[:,1]):.4f}"
                  for m in models]
})
display(summary)

# Bar chart comparison
accs = [accuracy_score(y_test,p)*100 for p in preds]
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(model_names, accs,
              color=['#3498DB','#E67E22','#9B59B6','#E74C3C'],
              edgecolor='black', zorder=3)
ax.bar_label(bars, fmt='%.1f%%', padding=3, fontsize=11)
ax.set_ylim(85, 102)
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
ax.grid(axis='y', linestyle='--', alpha=0.5, zorder=0)
plt.tight_layout()
plt.show()

# ── STEP 14: Predict on New Patient ──────────────────────────
print("\n" + "="*55)
print("  SINGLE PATIENT PREDICTION (DEMO)")
print("="*55)

# Using first test sample as a "new" patient
new_patient_scaled = X_test_pca[[0]]
actual_label = y_test.values[0]

pred_svm  = svm.predict(new_patient_scaled)[0]
pred_rf   = rf.predict(new_patient_scaled)[0]
pred_nn   = nn.predict(new_patient_scaled)[0]
pred_hyb  = hybrid.predict(new_patient_scaled)[0]

label_map = {0: '⚠️  MALIGNANT', 1: '✅ BENIGN'}

print(f"\n  Actual Diagnosis  : {label_map[actual_label]}")
print(f"  SVM Prediction    : {label_map[pred_svm]}")
print(f"  RF  Prediction    : {label_map[pred_rf]}")
print(f"  NN  Prediction    : {label_map[pred_nn]}")
print(f"  Hybrid Prediction : {label_map[pred_hyb]}")

print("\n✅ Hybrid ML Pipeline Complete!")
```
